```python
forest_importances = {}
for modeltype in ['imagery','context_prism_pt','all_prism_pt']:#models.keys():
    from sklearn.inspection import permutation_importance

    features = models[modeltype]
    print(modeltype)
    X = trainsub1[features]
    Y = trainsub1['wet']


    model = RandomForestClassifier(random_state=42,
                                   max_depth=mdepth,n_estimators = nest)#class_weight='balanced')
    model.fit(X, Y)
    
    start_time = time.time()
    result = permutation_importance(
        model, X, Y, n_repeats=10, random_state=42, n_jobs=4
    )
    elapsed_time = time.time() - start_time
    print(f"Elapsed time to compute the importances: {elapsed_time:.3f} seconds")

    forest_importances[modeltype] = pd.Series(result.importances_mean, 
                                              index=features)

```
